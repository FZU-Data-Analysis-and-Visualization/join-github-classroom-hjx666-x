```python
import pandas as pd 
data_dict = {
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
}
# 用字典创建DataFrame
df1 = pd.DataFrame(data_dict)
df1
```
