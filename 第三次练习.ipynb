### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。

   

2. 用透视表：行是 gender，值是 score，聚合函数为中位数。

```python

# 1. 对score列做标准化
score_mean = df1['score'].mean()
score_std = df1['score'].std()
df1['score_standardized'] = (df1['score'] - score_mean) / score_std
df1

```

```python
# 2. 透视表：行是gender，值是score，聚合函数为中位数
pivot_table_result = df1.pivot_table(
    index='gender',    
    values='score',    
    aggfunc='median'   
)
pivot_table_result
```
